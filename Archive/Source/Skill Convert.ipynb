```python
# Convert skills into machine learning training dataset
import pandas as pd
# Read the file
occ=pd.read_csv("../Data/occupation.csv")
task=pd.read_csv("../Data/task.csv")
occ.head()
```

```python
# Get the unique Skills to convert into column names and append Title
columns=task["Title"].unique().tolist()
#columns.append("Title")
len(columns)
```

```python
# Get the unique titles to convert into rows
#row=skills["Title"].unique()
```

```python

# Store all the skills for the specific title as a list
title="Chief Executives"

try:
    val=task.loc[(task["Title"]==title) & (task["Task Type"]=="Core") ,"Task"].tolist()
    print(val)
except:
    print("pass")
# Append the title to the new row and then append to result
```

```python
# Store the result as a list of lists
results=[]
# For each title find the Skills Required
for title in columns:
    #print(title)
    # Store all the skills for the specific title as a list
    new_row=[]

    try:
        val=task.loc[(task["Title"]==title) & (task["Task Type"]=="Supplemental") ,"Task"].tolist()
        new_row.append(val)
    except:
        new_row.append(0)
# Append the title to the new row and then append to result
    new_row.append(title)
    results.append(new_row)
```

```python
results
```

```python
# Create a data frame
DF=pd.DataFrame(results)
DF.head()
DF=DF.rename(columns={0: 'SupplementalTasks', 1: 'Title'})
#DF.to_csv("../Data/KnowledgeCleansed.csv",index=False)
testdf=DF.merge(occ,on="Title")
testdf.head()
```

```python
#DF=pd.read_csv("../Data/KnowledgeCleansed.csv")
testdf=testdf.set_index("Title")
testdf.head()
```
